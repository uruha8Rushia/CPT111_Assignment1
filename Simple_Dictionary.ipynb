# Project Title: Dictionary

Project Specification :
This dictionary is a service for everyone who are learning English or just simply want a dictionary inside computer. It is also designed to provide a user -friendly interface for managing a dictionary. There are three main features, reading, add words and searching for a certain word. This dictionary is based on a text file which mean all the words are read from a text file and display it at the terminal. When adding words, it will also write into the text file and save it.

Problem Analysis

There are 6 main function in my dictionary program, read, add new word, search, rearrange, word suggestion and stop.

1. Menu

*   Menu option must be shown after every operation like search and add word
*   Clear and interactive

2. Read

*   Display all available words from the input file "wordMeaning.txt"

3. Add word

*   User can add words to the text file
*   Initial of the word must be capital letter or the program will loop until user input correct format

4. Search word

*   Input a certain word to search in the "wordMeaning.txt" file
*   Initial of the word must be capital letter or the program will loop until user input correct format
*   Display the word and it's meaning to user
*   If the word is not found in the text file, a message will be displayed and suggest user to add it in our program

5. Rearrange in a new format

*   Output into "newWordMeaning.txt" file in a new format
*   Including index and the initial of the word
*   Example:
            30 tell t to communicate information or facts to someone.
            31 thing t an inanimate object or item.
            32 think t to have a particular opinion, belief, or idea.

6. Word suggestion

*   Display random word and it's meaning to user
*   User can learn a word

7. Remove word

*   Remove an unwanted word from the file at a time
*   User can remove the word he didn't like and remain the words he wants and interested

8. Stop

*   Stop the program
*   display appreciation to user

9. Comment

*   Comment is written in the source code to understand the part of the code more clearly

The implementation of the program is given in the following code.

```python
%%writefile Dictionary.cpp

#include <iostream>
#include <fstream>
#include <ctime>
#include <cstdlib>
#include <string>
using namespace std;

int main(){

  fstream myFile;		      //to select mode when opening file Ex; append mode
  ifstream inputFile;
  ofstream outputFile;
  string input, word, newWord, definition, search, compare, removeWord;
  int mode;
  bool isCapital = true, isSearched = false;
  char initial;
  unsigned seed = time(0);        //to get random everytime
  srand(seed);

  do{

      //menu option for user
      cout << "**************** DICTIONARY ********************" << endl;
      cout << "Menu Option\n";
      cout << "1. Read\n";     					    //display word available in dictionary
      cout << "2. Add a new word\n";  			//Add a new word
      cout << "3. Search a word\n";     		//search for a word
	    cout << "4. Word suggestion\n";       //display a random word for user
      cout << "5. Remove word\n";
      cout << "6. Stop\n";        				//stop or quit the program
      cout << "************************************************" << endl;

      cout << "Enter the mode you want: ";
      cin >> mode;

      if(mode < 1 || mode > 7){
      	cout << "Invalid input! Please choose the mode base on the menu.\n";    //display invalid input message to user when tested
	  }

      cin.clear();
      cin.ignore(10, '\n');       //to ignore invalid input like abcd if not it will occur repeatition

      switch (mode){

      case 1:
        inputFile.open("wordMeaning.txt");

        if(inputFile){
          while(getline(inputFile,input)){
              cout << input << endl;          //output to terminal all the words available in text file
          }
          inputFile.close();
        }
        else{
          cout << "Error! The file is not found.";
        }

        break;

      case 2:

          do{
          	cout << "Enter the word you want to add(Ex: Play): ";
            cin >> newWord;                          //user input new word to be added
          	outputFile.open("draft.txt");            //output to draft text file to check initial
          	if(outputFile){
          		outputFile << newWord;
          		outputFile.close();
			      }
          inputFile.open("draft.txt");               //read the draft text file to check initial
          if(inputFile){
            inputFile.get(initial);                  //get only one character
            if(initial < 'A' || initial > 'Z'){         //use char variable to check the initial is capitalised
              cout << "Invalid input. The initial of the word must be capital letter.\n";
              isCapital = false;
              inputFile.close();
            }
            else{
              inputFile.close();
              isCapital = true;
              break;
            }
          }
		  }while(!isCapital);            //a bool flag to make sure user enter valid input

          cin.ignore();										//clear buffer to use getline
          cout << "Enter the word's definition: ";
          getline(cin, definition);

          myFile.open("wordMeaning.txt", ios::app);			//use append mode to add a new word into the dictionary
          if(myFile.is_open()){
            myFile << newWord << "\t" << definition << endl;
            myFile.close();
          }
          break;

      case 3:

      	  do{
      	  	cout << "Enter the word you want to search(Ex: Play): ";   //Show user how to add new word
      	  	cin >> search;

      	  	outputFile.open("draft.txt");             //output to draft text file to check initial
          	if(outputFile){
          		outputFile << search;
          		outputFile.close();
			  }
			inputFile.open("draft.txt");                  //read the draft text file to check initial
			if(inputFile){
				inputFile.get(initial);                    //get only one character
				if(initial < 'A' || initial > 'Z'){                   //use char variable to check the initial is capitalised
					cout << "Invalid input. The initial of the word must be capital letter.\n";
					isCapital = false;
					inputFile.close();
				}
				else{
					inputFile.close();
					isCapital = true;
					break;
				}
			}
			}while(!isCapital);                 //a bool flag to make sure user enter valid input

      	  inputFile.open("wordMeaning.txt");

      	  if(inputFile){
      	  	while(getline(inputFile, compare, '\t')){		//loop to get the word to compare with user input
      	  		getline(inputFile, definition);           //take the definition so that can skip to next line
      	  		if(search != compare){
                isSearched = false;
      	  			continue;						//continue the loop when it is not the same word
					    }
				      else if(search == compare){
					      isSearched = true;
					      break;							//break through the loop and display the word
				      }
		        }
            if(isSearched){
              cout << search << "\t" << definition << endl;
            }
            else{
              cout << "The word did not contain in our file. Please consider adding it. Thank you so much\n";
              //message to user so that it can expand our word in dictionary
            }
          inputFile.close();
          }
          break;

	   case 4:
      int line, lines, randomLine;
	    inputFile.open("wordMeaning.txt");

        if(inputFile){
          line = 0;
          while(getline(inputFile,input)){			//to get the number of words inside dictionary
              line++;
          }
        }
        inputFile.close();

        randomLine = rand()% line + 1;				//select random line to be shown

	   	  inputFile.open("wordMeaning.txt");

        if(inputFile){
          lines = 0;
           while(lines != randomLine){				//loop through the word to get the selected word
              ++lines;
              getline(inputFile,input);
              }

          cout << "A word a day\n";
          cout << "*******************\n";			//show user the word and it's meaning
          cout << input << endl;
          cout << "Hope you have learn a new word today!\n";
          cout << "*******************\n";
          inputFile.close();
        }
	    break;

    case 5:
      	do{
      		cout << "Enter the word you want to remove(Ex: Play): ";   //Show user how to remove a word
      	  cin >> removeWord;

      	  outputFile.open("draft.txt");       //for checking initial
          if(outputFile){
          	outputFile << removeWord;
          	outputFile.close();
			    }
			   inputFile.open("draft.txt");
			   if(inputFile){
				 inputFile.get(initial);
				 if(initial < 'A' || initial > 'Z'){          //Check initial is capital letter or not
					 cout << "Invalid input. The initial of the word must be capital letter.\n";
					 isCapital = false;
					 inputFile.close();
				 }
				 else{
					 inputFile.close();
					 isCapital = true;
					 break;
				 }
			  }
		}while(!isCapital);       //loop until valid input

    inputFile.open("wordMeaning.txt");
    if(inputFile){
      	while(getline(inputFile, compare, '\t')){		//loop to get the word to compare with user input
      	  getline(inputFile, definition);           //take the definition so that can skip to next line
      	  if(removeWord != compare){
              isSearched = false;
      	      continue;						//continue the loop when it is not the same word
				  }//to find user's input word is inside the file
				  else if(removeWord == compare){
					    isSearched = true;
              inputFile.close();
					    break;							//break through the loop
				  }
		    }
    inputFile.close();

		inputFile.open("wordMeaning.txt");
		outputFile.open("draft.txt");
		if(inputFile){
      if(isSearched){
        while(inputFile >> word){
				  getline(inputFile, definition);
				  if(removeWord != word){                             //if not the user's input word then output to draft
					  outputFile << word << "\t" << definition << endl;
				  }
				  else if(removeWord == word){                       //the user's input word will be skipped
					  continue;
				  }
			  }
        cout << "The word have been removed\n";
      }
      else{
        cout << "The word did not contain in our file. Please consider adding it. Thank you so much\n";
      }
		}
		inputFile.close();
		outputFile.close();

		inputFile.open("draft.txt");              //output to the original file
		outputFile.open("wordMeaning.txt");
		if(inputFile){
          while(getline(inputFile,input)){
              outputFile << input << endl;
          }
          inputFile.close();
          outputFile.close();
        }
    }
		break;

    case 6:
      cout << "Thanks for visiting! ^^\n";		//appreciation and quit from dictionary
      break;

    }
  }while(mode != 6);

  inputFile.open("wordMeaning.txt");
  outputFile.open("newWordMeaning.txt");			//output into a another file with new format

  if(inputFile){
    int index = 0;
    while(inputFile.get(initial)){				//only taken the initial of the word
      inputFile >> word;                  //take the rest of the word and stopped when hits empty spaces
      getline(inputFile, definition);
      word = initial + word;					//add back the initial into the word
      index++;
      outputFile << index << " " << word << " " << initial << definition << endl;     //output to another file in another format
		}

		inputFile.close();
		outputFile.close();
		}

    inputFile.open("newWordMeaning.txt");			//display the changes

		if(inputFile){
			while(getline(inputFile, input)){        //output to user the new format with index and initial
				cout << input << endl;
			}
			inputFile.close();
		}

  return 0;
}
```
